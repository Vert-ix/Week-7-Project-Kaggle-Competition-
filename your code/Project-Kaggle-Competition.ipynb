```python
df.job_title = ["Data Engineer" if "Engineer" in word else word for word in df.job_title]
df.job_title = ["Data Architect" if "Architect" in word else word for word in df.job_title]
df.job_title = ["Data Analyst" if "Analyst" in word or "Data Analytics Lead" in word else word for word in df.job_title]
df.job_title = ["Computer Vision" if "ision" in word else word for word in df.job_title]
df.job_title = ["Data Scientist" if "Data Scientist" in word or "Data Science" in word else word for word in df.job_title]
df.job_title = ["Machine Learning Engineer" if "Machine Learning Engineer" in word or "Machine Learning Developer"  in word or "Machine Learning Manager" in word or "Machine Learning Engineer"  in word else word for word in df.job_title]
df.job_title = ["Machine Learning Scientist" if "Machine Learning Scientist" in word else word for word in df.job_title]
df.job_title = ["Finance Data Analyst" if "Finance" in word or "Financial" in word else word for word in df.job_title]

top7 = list(df.job_title.value_counts().index[:9])

df.loc[-df.job_title.isin(top7), ['job_title']] = 'Other' 

df.job_title.value_counts()
```
